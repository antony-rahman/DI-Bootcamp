```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

Rows - iloc (index location)

1 Row

```python
df.iloc[0]
```

Multiple rows

```python
df.iloc[0:5]
```

```python
df.iloc[50:60]
```

Multiple rows with step

```python
df.iloc[50:100:3]
```

Condition based indexing - loc (locate)

```python
df['species'] == 'setosa'
```

```python
search_indexes = [1,5,6]
```

```python
df.loc[search_indexes]
```

```python
df.loc[df['species'] == 'setosa'] 
```

What indexing types we have in pandas?
    - iloc - when we know the index
    - loc - when we know the condition

Columns

One column

```python
df['species'].iloc[30:50]
```

Multiple columns

```python
df[['species', 'petal_width']]
```

```python
df[['species', 'petal_width']].iloc[:10]
```

Filtering - df [cond]

One column filtering

```python
df[df['species'] == 'setosa']
```

```python
df[df['sepal_width'] >= 3.5]
```

Multiple column filtering

& = and 

| = or

< = smaller

'>' = higher

!= - not equal

```python
df[(df['species'] == 'setosa') & (df['sepal_width'] >= 3.5)]
```
